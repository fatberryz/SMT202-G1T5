```python
#soup = BeautifulSoup(page_source, 'html.parser')
next_page = "//*[@class='ui_button nav next primary ']"
page_down = "window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;"
domain = 'https://www.tripadvisor.com.sg'

# username_list = []
# num_posts_list = []
# date_posted_list = []
# date_experienced_list = []
# rating_list = []
# review_title_list = []
# review_text_list = []

with open('asian_civilization_musueum_tripadvisor.csv', 'a', encoding = "utf-8") as csvfile:
    fieldnames = ['username', 'num_posts', 'date_posted', 'date_experienced', 'rating', 'review_title', 'review_text']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    index = 0
    
    for p in range(0, 355): #hardcoded - eyeball value now
        try:
            print("Scraping Page " + str(p+1) + "\nCurrent Time:" + str(dt.datetime.now()))
            more_buttons = driver.find_elements_by_class_name("location-review-review-list-parts-ExpandableReview__cta--2mR2g")
            for x in range(len(more_buttons)):
                if more_buttons[x].is_displayed():
                    driver.execute_script("arguments[0].click();", more_buttons[x])
                    time.sleep(2)
                    break
        except StaleElementReferenceException as e:
            raise e
            
            
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        soup_review_blocks = soup.find_all('div', class_='location-review-card-Card__ui_card--2Mri0 location-review-card-Card__card--o3LVm location-review-card-Card__section--NiAcw')

    #Select all review blocks on current page
    #soup_review_blocks = soup.find_all('div', class_='location-review-card-Card__ui_card--2Mri0 location-review-card-Card__card--o3LVm location-review-card-Card__section--NiAcw')

        for i in range(len(soup_review_blocks)):
            #get username
            usernames = soup_review_blocks[i].find_all("a", class_="ui_header_link social-member-event-MemberEventOnObjectBlock__member--35-jC")
            username_list = usernames[0].children
            for username in username_list:
                username = username
            #time.sleep(2)
            #username_list.extend(username[0].children)

            #get number of posts user has
            num_posts = soup_review_blocks[i].find_all("span", class_="social-member-MemberHeaderStats__bold--3z3qh")
            num_post_list = num_posts[0].children
            for num_post in num_post_list:
                num_post = num_post
            #num_posts_list.extend(num_posts[0].children)

            #get date of review post
            review_date = soup_review_blocks[i].find_all("a", class_="ui_header_link social-member-event-MemberEventOnObjectBlock__member--35-jC")
            review_date_full = str(review_date[0].next_sibling)
            review_date = review_date_full.partition("wrote a review ")[2]
            #date_posted_list.append(date.partition("wrote a review ")[2])

            #get date of experience
            experience_dates = soup_review_blocks[i].find_all("span", class_="location-review-review-list-parts-EventDate__bold--3b3PE")
            if len(experience_dates) == 0:
                experience_date = "NA"
            else:
                experience_date_full = str(experience_dates[0].next_sibling)
                experience_date = experience_date_full.strip()
            #time.sleep(2)
            #date_experienced_list.append(date)

            #get rating (rating out of 5 stars)
            ratings = soup_review_blocks[i].find_all("div", class_="location-review-review-list-parts-SingleReview__mainCol--1hApa")
            rating_parent = ratings[0].find_all("div", class_= "location-review-review-list-parts-RatingLine__bubbles--GcJvM")
            rating_whole = rating_parent[0].children
            for item in rating_whole:
                item = str(item)
                ratings = re.findall('[0-9][0-9]', item)
                rating = ratings[0]
                #rating_list.append(rating[0])

            #get review title
            titles = soup_review_blocks[i].find_all("a", class_="location-review-review-list-parts-ReviewTitle__reviewTitleText--2tFRT")
            if len(titles) == 0:
                title = "NA"
            else:
                for title in titles[0].children:
                    title = str(title)
                    titles = re.findall('<span>(.+?)</span>', title)
                    if len(titles) == 0:
                        title = "NA"
                    else:
                    #title = titles[0]
                        title = titles[0].split(">")[1]
                    #review_title_list.append(title)
            time.sleep(2)

            #get review text
            texts = soup_review_blocks[i].find_all("q", class_="location-review-review-list-parts-ExpandableReview__reviewText--gOmRC")
            for text in texts[0].children:
                text = str(text)
                texts = re.findall('<span>(.+?)</span>', text)
                text = texts[0]
                #review_text_list.append(texts[0])
                
                
            writer.writerow(
                {
                    'username':username,
                    'num_posts':num_post,
                    'date_posted':review_date,
                    'date_experienced':experience_date,
                    'rating':rating,
                    'review_title':title,
                    'review_text':text
                }
               )
        
        try:
            driver.execute_script(page_down)
            time.sleep(2)
            #driver.find_element_by_xpath(next_page).click()
            #time.sleep(8)
            more_pages = driver.find_elements_by_xpath(next_page)
            for x in range(len(more_pages)):
                if more_pages[x].is_displayed():
                    driver.execute_script("arguments[0].click();", more_pages[x])
                    time.sleep(5)
                    break
        except:
            print('end of scraping page', p)
# print(username_list)
# print(num_posts_list)
# print(date_posted_list)
# print(date_experienced_list)
# print(rating_list)
# print(review_title_list)
# print(review_text_list)
    
```
