```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('linkedinprofiles_raw.csv')
df.head()
```

### Drop columns that have too many missing values

```python
print("Before dropping:",len(df.columns),"columns",'\n')
#drop columns with more than 80% of values missing
df_dropped = df.dropna(thresh = (len(df) * .2) , axis = 1)

print("List of columns dropped:")
print([col for col in df.columns.difference(df_dropped.columns)],'\n')

df = df_dropped
print("After dropping:",len(df.columns),"columns")
```

```python
modes = ['Previous Arts CCA_Performance','Year_Year 2','Year_Year 1','School_School of Accountancy','School_School of Economics']
means = ['Participation in art activities','Career oriented']
df = df[df['Education 1'].str.lower().str.contains("singapore management university")]
```

```python
#clean and map job titles
#education description get their cca, exco or no, deanlist/scholarship
df[['Education 1',
       'Education Degree 1', 'Education FOS 1', 'Education Start 1',
       'Education End 1', 'Education Description 1', 'Education 2',
       'Education Degree 2', 'Education FOS 2', 'Education Start 2',
       'Education End 2', 'Education Description 2', 'Education 3',
       'Education Degree 3', 'Education FOS 3', 'Education Start 3',
       'Education End 3', 'Education Description 3']]
```

#### Extract students who have studied Accountancy or Economics

```python
df = df[~df['Education Degree 1'].isna()]
df['School_School of Accountancy'] = 0
df['School_School of Accountancy'][df['Education Degree 1'].str.lower().str.contains("accountancy")] = 1

df['School_School of Economics'] = 0
df['School_School of Economics'][df['Education Degree 1'].str.lower().str.contains("economics")] = 1
```

```python
df['Year_Year 1'] = 0
df['Year_Year 1'][df['Education Start 1'] == 2019] = 1

df['Year_Year 2'] = 0
df['Year_Year 2'][df['Education Start 1'] == 2018] = 1
```

#### To get an idea of their participation in art activities, we can seek out the following:
- Participated in any art CCA or activities
- Exco for any art CCA or activities

#### We can also look at their previous education descriptions to determine if they have participated in arts-related CCAs before


```python
art_clubs = ['ACF','Ardiente','Artdicted','Ballare','Broadcast & Entertainment','Caderas Latinas','Chamber Choir','Chinese Orchestra','Eurhythmix','Funk Movement','Guitarissimo','INDANCITY','Indian Cultural Society','Ivory Keys','Komunitas Indonesia','Literati','Malay Language & Cultural Club','SMUSAIC','Samba Masala','SoundFoundry','StageIT','Stereometa','Symphonia','Voix']
```

```python
#Get Arts CCAs
edu_df = df.loc[:, df.columns.str.startswith('Education Description 1')]
edu_df = edu_df[edu_df.apply(lambda r: r.str.contains('|'.join(art_clubs), case=False).any(), axis=1)]
CCA_ind = edu_df.index.tolist()
```

```python
#Get Art activities
edu_df = df.loc[:, df.columns.str.startswith('Education Description')]
edu_df = edu_df[edu_df.apply(lambda r: r.str.contains(' art | arts ', case=False).any(), axis=1)]
Activity_ind = edu_df.index.tolist()
```

```python
#Get Art exco/significant positions
#Assume all records here show significant achievements like exco or director for some committee so we give extra weights to that
org_ind = []
for i in range(1,4):
    org_df = df.loc[:, df.columns.str.contains('Organization '+str(i)+'|Organization Title '+str(i),regex=True)]
    org_ind += org_df[org_df.apply(lambda r: r.str.contains('|'.join(art_clubs), case=False).any(), axis=1)].index.tolist()
```

```python
from collections import Counter
data = Counter(CCA_ind+Activity_ind+org_ind)
total_ind = [ind for ind, score in data.most_common()]
total_score = [score for ind, score in data.most_common()]

#Count those indexes, then give score based on their index, then scale from 1 to 4
df['Participation in art activities'] = 0 
df['Participation in art activities'].loc[total_ind] = total_score
```

#### Get previous arts cca with performance or sculpture

```python
art_list = ['orchestra','dance','choir','drama','band','ballet','theatre']
prev_df = df.loc[:, df.columns.str.contains('Education Description 2|Education Description 3')]
prev_df = prev_df[prev_df.apply(lambda r: r.str.contains('|'.join(art_list), case=False).any(), axis=1)]
prev_ind = prev_df.index.tolist()
```

```python
data = Counter(prev_ind)
total_ind = [ind for ind, score in data.most_common()]
total_score = [score for ind, score in data.most_common()]

df['Previous Arts CCA_Performance'] = 0
df['Previous Arts CCA_Performance'].loc[total_ind] = total_score
```

#### Know how career oriented they are by looking at how many internships completed and time spent working there

```python
#If organization title cntains "intern" then we count as 1 intern completed
car_ind = []
length_intern = []
ind_intern = []
df['Organization End 1'][df['Organization End 1'] == "PRESENT"] = "Mar-20"


for i in range(1,8):
    car_df = df.loc[:, df.columns.str.contains('Organization Title '+str(i),regex=True)]
    curr_ind = car_df[car_df.apply(lambda r: r.str.contains('intern', case=False).any(), axis=1)].index.tolist()
    car_ind += curr_ind
    
    df['Organization Start '+str(i)][df['Organization Start '+str(i)]=='2020'] = '01-20'
    df['Organization Start '+str(i)][df['Organization Start '+str(i)]=='2018'] = '01-18'
    df['Organization Start '+str(i)][df['Organization Start '+str(i)]=='2015'] = '01-15'
    df['Organization Start '+str(i)][df['Organization Start '+str(i)]=='2019'] = '01-19'
    df['Organization Start '+str(i)][df['Organization Start '+str(i)]=='2017'] = '01-17'   
    df['Organization Start '+str(i)][df['Organization Start '+str(i)]=='2016'] = '01-16'   
    df['Organization End '+str(i)][df['Organization Start '+str(i)]=='2016'] = '01-16'   
    df['Organization End '+str(i)][df['Organization End '+str(i)]=='2020'] = '01-20'
    df['Organization End '+str(i)][df['Organization End '+str(i)]=='2018'] = '01-18'
    df['Organization End '+str(i)][df['Organization End '+str(i)]=='2015'] = '01-15'
    df['Organization End '+str(i)][df['Organization End '+str(i)]=='2019'] = '01-19'
    df['Organization End '+str(i)][df['Organization End '+str(i)]=='2017'] = '01-17'  
    df['Organization End '+str(i)][df['Organization Start '+str(i)]=='2016'] = '01-16'  
    df['Organization End '+str(i)][df['Organization End '+str(i)]=='2016'] = '01-16'  
    df['Organization End '+str(i)][df['Organization Start '+str(i)]=='2012'] = '01-12'  
    df['Organization End '+str(i)][df['Organization End '+str(i)]=='2012'] = '01-12'  
    
    ind_intern.append(curr_ind)
    length_intern.append((df.loc[curr_ind]['Organization Start '+str(i)].fillna('0-').tolist(),df.loc[curr_ind]['Organization End '+str(i)].fillna('0-').tolist()))
```

```python
data = Counter(car_ind)
total_ind = [ind for ind, score in data.most_common()]
total_score = [score for ind, score in data.most_common()]

df['Number of internships'] = 0
df['Number of internships'].loc[total_ind] = total_score
```

```python
#Time spent working there
#Scale value

month_dict = {'Jan':1,
             'Feb':2,
             'Mar':3,
             'Apr':4,
             'May':5,
             'Jun':6,
             'Jul':7,
             'Aug':8,
             'Sep':9,
             'Oct':10,
             'Nov':11,
             'Dec':12,
             'less than a year': 6,
             'PRESENT':3,
             '0' : 0,
             '01':1,
             '2016':0,
             '2014':0,
             '2012':0,
             '2013':0,
             '2011':0,
             '2010':0}

duration_intern = [month_dict[str(in_date).split('-')[0]] - month_dict[str(tup[1][ind]).split('-')[0]] for tup in length_intern for ind,in_date in enumerate(tup[0]) ]
```

```python
ind_intern = [x for i in ind_intern for x in i]
duration_df = pd.DataFrame([ind_intern,[abs(d) for d in duration_intern]]).T
duration_df.columns = ['index','month']
duration_df = duration_df.groupby('index').agg({'month':'sum'}).reset_index()
```

```python
df['Length of intern'] = 0
df['Length of intern'].loc[duration_df['index'].tolist()] = duration_df['month'].tolist()
```

#### Scale all relevant columns

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
df['Participation in art activities'] = scaler.fit_transform(df['Participation in art activities'].values.reshape(-1, 1))

scaler = MinMaxScaler(feature_range=(0,1.5))
df['Number of internships'] = scaler.fit_transform(df['Number of internships'].values.reshape(-1, 1)) #0 to 3
df['Length of intern'] = scaler.fit_transform(df['Length of intern'].values.reshape(-1, 1)) #0 to 3
```

##### Get a df with all relevant columns

```python
df['Career oriented'] = df['Length of intern'] + df['Number of internships']
df.to_csv('Linkedin_features_extracted.csv',index=False)
```
